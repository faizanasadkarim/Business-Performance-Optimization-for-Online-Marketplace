#### 1b. Understanding how many new customers acquired every month

```python
# Convert order purchase timestamp to datetime
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])
```

```python
# Merge customer_orders and orders on 'order_id' to get all relevant columns
merged_df = pd.merge(customers, orders, on='customer_id')

# Convert date columns to datetime if they aren't already
merged_df['order_purchase_timestamp'] = pd.to_datetime(merged_df['order_purchase_timestamp'])

# Extract year and month from the 'order_purchase_timestamp'
merged_df['year-month'] = merged_df['order_purchase_timestamp'].dt.to_period('M')
orders['year-month'] = orders['order_purchase_timestamp'].dt.to_period('M')


# Step 1: Group by customer_id, year, and month to get the minimum transaction date
min_monthly_transaction_df = merged_df.groupby(['customer_unique_id', 'year-month']).agg(
    min_transaction_date_month=('order_purchase_timestamp', 'min')
).reset_index()

# Step 2: Group by customer_id to get the first transaction date ever
min_overall_transaction_df = merged_df.groupby('customer_unique_id').agg(
    min_transaction_date=('order_purchase_timestamp', 'min')
).reset_index()

# Step 3: Merge the two dataframes on customer_id
customer_status_df = pd.merge(min_monthly_transaction_df, min_overall_transaction_df, on='customer_unique_id', how='left')

# Step 4: Create a new column to categorize customers as NEW or EXISTING
customer_status_df['customer_category'] = customer_status_df.apply(
    lambda row: 'EXISTING' if row['min_transaction_date_month'] > row['min_transaction_date'] else 'NEW', axis=1
)

customer_status_df=customer_status_df.drop_duplicates(subset="customer_unique_id",keep="last")

customers=pd.merge(customers,customer_status_df[["customer_unique_id","customer_category"]],on='customer_unique_id',how='left')
customers.head()
```

#### 1c. Understand the retention of customers on month on month basis


```python
data = pd.merge(orders, customers, on='customer_id')

# Convert timestamps to datetime format
data['order_purchase_timestamp'] = pd.to_datetime(data['order_purchase_timestamp'])

# Extract the month of the first purchase (Cohort Month)
data['cohort_month'] = data.groupby('customer_unique_id')['order_purchase_timestamp'].transform('min').dt.to_period('M')

# Extract the order month (Activity Month)
data['order_month'] = data['order_purchase_timestamp'].dt.to_period('M')

# Calculate the time offset (in months) between the cohort month and the order month
data['cohort_index'] = (data['order_month'] - data['cohort_month']).apply(lambda x: x.n)

# Count the number of unique customers for each cohort and cohort index
cohort_data = data.groupby(['cohort_month', 'cohort_index']).agg({'customer_unique_id': 'nunique'}).reset_index()

# Pivot the data to create a cohort table
cohort_table = cohort_data.pivot_table(index='cohort_month', columns='cohort_index', values='customer_unique_id')

# Calculate retention rates by dividing by the number of customers in each cohort's first month
cohort_size = cohort_table.iloc[:, 0]
retention = cohort_table.divide(cohort_size, axis=0)

# Visualize the retention table
plt.figure(figsize=(12, 8))
sns.heatmap(retention, annot=True, fmt='0.00%', cmap='Blues')
plt.title('Customer Retention Cohort Analysis (Month on Month)')
plt.xlabel('Months after First Purchase')
plt.ylabel('Cohort Month')
plt.show()
```

#### 3. Cross-Selling (Which products are selling together)

```python
from itertools import combinations
from collections import Counter

# Group by order_id and aggregate product_ids into a list
order_products = order_items.groupby('order_id')['product_id'].apply(list)

# Generate combinations of products for each order
comb_list = []
for prod in order_products:
    comb_list.extend(combinations(products, 2))  # Change 2 to 3 for triplets

comb_list2 = []
for prod in order_products:
    comb_list.extend(combinations(products, 3))
comb_list=comb_list+comb_list2  

# Count the frequency of each combination
comb_count = Counter(comb_list)

# Get the top 10 combinations
top_combinations = comb_count.most_common(10)

# Display the results
for comb, count in top_combinations:
    print(f"Combination: {comb}, Count: {count}")
```
